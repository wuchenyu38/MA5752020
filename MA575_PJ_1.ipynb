```
#Monthly Rebalancing
#initial total: 150000
#each: 10000
#create list to store monthly weights
#01/01/2012 as initial weight, rebalancing at each the end of months
#order F,MSFT,DIS,AAPL,AEP,GOOG,JPM,MCD,TIF,WMT,AMZN,GE,SPY,VT,VTI
```

```
def Rebalance(c,emp1,emp2):
  for x in range(2012,2016):
    f1=c[c.index.year==x]
    f1.index.to_pydatetime()
    for y in range(1,13):
      f2=f1[f1.index.month==y]
      emp1.append(f2.values.tolist()[-1])

  for x in range(2016,2020):
    f1=c[c.index.year==x]
    f1.index.to_pydatetime()
    for y in range(1,13):
      f2=f1[f1.index.month==y]
      emp2.append(f2.values.tolist()[-1])
```

```
#Ford Rebalancing Prices
import datetime as datetime

Ff = F.to_frame()
FRc=[] #monthly rebalancing price for calibration
FRb=[] #monthly rebalancing price for back-testing
Ff.index.to_pydatetime()
Rebalance(Ff,FRc,FRb)
```
